```python
import pandas as pd
import numpy as np

# Get data from csv
reviews = pd.read_csv('datasets/vivino-reviews.csv', sep=',')
```

```python
from collections import Counter #single word

res = Counter(" ".join(reviews["note"]).split()).most_common(300)

result = []
for index,r in enumerate(res):
    found = False
    for i in range(index+1,300):
        if r[0].lower() == res[i][0][:-1].lower():
            result.append((r[0].lower(), r[1]+res[i][1])) 
            found = True
            break
        elif r[0].lower() == res[i][0].lower():
            result.append((r[0].lower(), r[1]+res[i][1])) 
            found = True
            break
        elif len(r[0]) <= 2:
            found = True
            break

    if not(found):
        result.append((r[0],r[1]))

for r in result:
    print(r[0],",",r[1])
```

```python
from collections import Counter  #bigrams

res = " ".join(reviews["note"]).split()
res = list(res)

bigrams = zip(res, res[1:])
counts = Counter(bigrams)
res = counts.most_common(200)

for r in res:
    print(r[0][0],r[0][1],",",r[1])
```

```python
def find_ngrams(input_list, n):
  return zip(*(input_list[i:] for i in range(n)))

res = " ".join(reviews["note"]).split()
res = list(res)

trigrams = find_ngrams(res, 3)
counts = Counter(trigrams)
res = counts.most_common(200)

for r in res:
    print(r[0][0],r[0][1],r[0][2],",",r[1])
```
