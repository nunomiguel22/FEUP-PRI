```python
import pandas as pd
import numpy as np

# Get data from csv
reviews = pd.read_csv('vivino-reviews.csv', sep=',')

reviews
```

```python
from collections import Counter

res = Counter(" ".join(reviews["note"]).split()).most_common(300)

result = []
for index,r in enumerate(res):
    found = False
    for i in range(index+1,300):
        if r[0].lower() == res[i][0][:-1].lower():
            result.append((r[0].lower(), r[1]+res[i][1])) 
            found = True
            break
        elif r[0].lower() == res[i][0].lower():
            result.append((r[0].lower(), r[1]+res[i][1])) 
            found = True
            break
        elif len(r[0]) <= 2:
            found = True
            break

    if not(found):
        result.append((r[0],r[1]))

for r in result:
    print(r[0],",",r[1])
```
